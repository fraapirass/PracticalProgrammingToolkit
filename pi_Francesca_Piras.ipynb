```python
import math as mh
import random as rnd
import numpy as np
```

```python
N = 500000
M = 0

N_oa = N
M_oa = 0

err = np.sqrt(N)/100
err_oa = np.sqrt(N_oa)/100

dots = []
dots_oa = []
```

```python
for i in range(N) : #loop su un quadrante
    
    x = rnd.uniform(0, 1)
    y = rnd.uniform(0, 1)
    
    dots.append([x, y])

dots_arr = np.array(dots)
```

```python
for i in range(N_oa) : #loop su tutto il quadrato
    
    x_oa = rnd.uniform(-1, 1)
    y_oa = rnd.uniform(-1, 1)
    
    dots_oa.append([x_oa, y_oa])

dots_oa_arr = np.array(dots_oa)
```

```python
for [x, y] in dots_arr[ 0 : -1 ] :

    if x*x + y*y <= 1 :
        M = M + 1

for [x_oa, y_oa] in dots_oa_arr[ 0 : -1 ] :

    if x_oa*x_oa + y_oa*y_oa <= 1 :
        M_oa = M_oa + 1
```

```python
pi = 4 * M / N
oa_pi =  4 * M_oa / N_oa

print( pi , oa_pi )
print('discrepanza %: ', abs(pi - np.pi)*100/np.pi, abs(oa_pi - np.pi)*100/np.pi)
```
